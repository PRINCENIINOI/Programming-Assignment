```python
import matplotlib.pyplot as plt
```

```python
fileName = "./red_wine.csv"
redWine = pd.read_csv(fileName)
```

```python
print(redWine.describe())
```

```python
#Histogram
plt.figure(figsize=(10, 10))
tableofContent=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']
for x, content in enumerate(tableofContent):
    plt.subplot(6,4,x+1)
    plt.hist(redWine[content].dropna(),bins=50,alpha=1)
    plt.title(content.replace("_"," ").title())
    plt.tight_layout()
plt.show()

```

```python
# Scatter plot
plt. figure (figsize=(8, 6)) 
plt. scatter(redWine['alcohol'],redWine['quality'],alpha=1,color='blue')
plt. title( 'Amount of Alcohol/Quality of wine')
plt. xlabel( 'Amount of Alcohol' ) 
plt. ylabel('Quality of wine')
plt. grid (True)
plt. show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(redWine['alcohol'],vert=False)
plt.xlabel('Amount of Alcohol')
plt.title( 'Chane In Amount of Alcohol')
plt. show()
```

# Report on Graphs
### HISTOGRAM
The histogram shows the amount of each chemical in the wine data given.

### Box plot 
The box plot shows the change in the amount of alcohol. The horizontal line inside the box represents the median alcohol content.

### Scatter plot 
The scatter plot shows the relationship between amount of alcohol and quality of wine.The x-axis represents the amount of alcohol and the y-axis represents the quality of the wine. It has been observed that wine which contains more alcohol is most likely to be of better quality.
